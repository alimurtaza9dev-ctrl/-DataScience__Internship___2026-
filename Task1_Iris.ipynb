# Task 1: Exploring and Visualizing the Iris Dataset

**Objective:**  
Understand the structure of the Iris dataset and visualize its features to gain insights.

**Dataset:**  
The Iris dataset contains 150 samples of iris flowers, with 4 features (sepal length, sepal width, petal length, petal width) and the species label.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style for better visuals
sns.set(style="whitegrid")
```

```python
# Load Iris dataset from seaborn
iris = sns.load_dataset('iris')

# Display first 5 rows
iris.head()
```

```python
# Check dataset shape
print("Dataset shape:", iris.shape)

# Check column names
print("Columns:", iris.columns)

# Summary statistics
print(iris.describe())

# Check for missing values
print("Missing values:\n", iris.isnull().sum())
```

```python
# Pairplot to visualize relationships between features
sns.pairplot(iris, hue='species')
plt.show()
```

```python
# Histogram for each numeric feature
iris.hist(figsize=(10,8), bins=15)
plt.show()
```

```python
# Box plot to check for outliers
plt.figure(figsize=(10,6))
sns.boxplot(data=iris)
plt.show()
```

# Conclusion

- The Iris dataset has 150 rows and 5 columns (4 features + species).  
- Scatter plots show clear separation between species for petal features.  
- Histograms show the distribution of each feature.  
- Box plots help detect outliers and understand the spread of the data.  

This completes basic exploration and visualization of the Iris dataset.
